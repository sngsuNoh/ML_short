# 네이버 블로그 검색결과 크롤링 (2)

```python
from bs4 import BeautifulSoup
import requests
```

```python
base_url = 'https://search.naver.com/search.naver?ssc=tab.blog.all&sm=tab_jum&query='
```

```python
keyword = input()
```

```python
search_url = base_url + keyword
search_url
```

```python
r = requests.get(search_url)
r.text
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
soup
```

```python
items = soup.select('.view_wrap')
items
```

```python
for rank, item in enumerate(items, 1):
    post_title = item.select_one('.title_link')
    ad = item.select('.link_ad')
    if ad:
        print(f'{rank} : 광고') 
        continue
    print(f'{rank} : {post_title.text}')
```

```python
for rank, item in enumerate(items, 1):
    print(f'{rank}')
    if item.select('.link_ad'): 
        print('광고\n')
        continue
    blog_title = item.select_one('.name')
    post_title = item.select_one('.title_link')
    print(f'{blog_title.text}')
    print(f'{post_title.text}')
    print(f'{post_title.get('href')}')
    print()
```
